```python
import pydeck as pdk
import pandas as pd
import numpy as np

# Generate 3D point cloud data
N = 1000
points = np.random.randn(N, 3) * 10  # Scale to make visible
df = pd.DataFrame(points, columns=["x", "y", "z"])

# Use pydeck's PointCloudLayer (no map background)
layer = pdk.Layer(
    "PointCloudLayer",
    data=df,
    get_position="[x, y, z]",
    point_size=5,
)

# Center the view around the cloud (assume origin-centered cloud)
view_state = pdk.ViewState(
    longitude=0, latitude=0, zoom=1,
    pitch=60, bearing=0
)

# Compose and display deck without map
deck = pdk.Deck(
    layers=[layer],
    initial_view_state=view_state,
    map_provider=None,  # <--- No map!
)
deck.show()
```
