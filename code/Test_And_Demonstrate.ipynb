```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift, estimate_bandwidth
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate sample data
nums = np.arange(0, 11, 2)
centers = np.array(np.meshgrid(nums, nums)).T.reshape(-1, 2)

clusterCenters = []

minimum = 0
maximum = 5

for i in range(20):
    print("I:", i)
    X, _ = make_blobs(n_samples=len(centers) * 50, centers=centers, cluster_std=0.4)

    window = (maximum - minimum) / 2
    bandwidth = window

    while True:
        window /= 2
        
        ms = MeanShift(bandwidth=bandwidth)
        ms.fit(X)
        # print("len(ms.cluster_centers_):", len(ms.cluster_centers_))

        if (len(ms.cluster_centers_) < len(centers)):
            bandwidth -= window
        elif (len(ms.cluster_centers_) > len(centers)):
            bandwidth += window
        else:
            break
    
    clusterCenters.append(ms.cluster_centers_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=ms.labels_, cmap='viridis', s=30)
plt.scatter(ms.cluster_centers_[:, 0], ms.cluster_centers_[:, 1], c='red', s=200, marker='X')
plt.title("Mean Shift Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
plt.clf()
```

```python
# Plot the results
plottableCenters = np.array([pair for sublist in clusterCenters for pair in sublist])

plt.figure(figsize=(8, 6))
plt.scatter(plottableCenters[:, 0], plottableCenters[:, 1], c='red')
plt.title("Mean Shift Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
kmeanClusterCenters = []

for i in range(20):
    print("I:", i)

    # Generate data around true centers
    X, _ = make_blobs(n_samples=len(centers) * 50, centers=centers, cluster_std=0.4)

    # Use KMeans with known number of clusters
    kmeans = KMeans(n_clusters=len(centers), init='k-means++')
    kmeans.fit(X)

    # Save the detected centers
    kmeanClusterCenters.append(kmeans.cluster_centers_)
```

```python
# Plot the results
kmeanPlottableCenters = np.array([pair for sublist in kmeanClusterCenters for pair in sublist])

plt.figure(figsize=(8, 6))
plt.scatter(kmeanPlottableCenters[:, 0], kmeanPlottableCenters[:, 1], c='red')
plt.title("K-Means")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```
